```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rv_bernoulli = stats.bernoulli(p=0.3)
```

```python
rv_bernoulli.rvs(14)
```

```python
rv_binom = stats.binom(100, p=0.9)
```

```python
rv_binom.rvs(8)
```

```python
plt.hist(rv_binom.rvs(80), bins=10)
```

```python
a = 5
b = 10
rv_uniform = stats.uniform(a, b - a)
```

```python
rv_uniform.cdf(5.5)
```

```python
rv_uniform.pdf(5.5)
```

```python
X = np.linspace(a - 2, b + 2, 100)
cdf = rv_uniform.cdf(X)
plt.plot(X, cdf)
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Cumulative distribution function for uniform')
```

```python
pdf = rv_uniform.pdf(X)
plt.plot(X, pdf)
plt.xlabel('X')
plt.ylabel('f(x)')
plt.title('Probability distribution function for uniform')
```

```python
mu=2
sigma=0.5
rv_norm = stats.norm(loc=mu, scale=sigma)
rv_norm.rvs(17)
```

```python
X = np.linspace(0, 4, 100)
norm_cdf = rv_norm.cdf(X)
plt.plot(X, norm_cdf)
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Cumulative distribution function for normal')
```

```python
norm_pdf = rv_norm.pdf(X)
plt.plot(X, norm_pdf)
plt.xlabel('X')
plt.ylabel('f(x)')
plt.title('Probability distribution function for normal')
```
