```python
import pandas as pd
from numpy.random import exponential
```

```python
df = pd.DataFrame({'x': range(20), 'y': exponential(10, 20)})
```

```python
%matplotlib inline
```

```python
df.y.hist()
```

```python
import matplotlib.pyplot as plt
```

```python
data1 = exponential(5, 10)
data2 = exponential(6, 20)
```

```python
plt.plot(data1)
```

```python
plt.scatter(range(len(data2)), data2)
```

```python
plt.plot(data1, label='Первый игрок')  # непрерывный график
plt.scatter(range(len(data2)), data2, color='red', label='Второй игрок')  # график точек
plt.title('Результаты', fontdict={'fontsize': 20})  # заголовок
plt.xlabel('Номер попытки')  # подпись оси х
plt.ylabel('Результат')  # подпись оси y
plt.legend()  # подпись графиков
plt.xticks(range(0,20,4))  # шаг осей
# plt.show()
plt.savefig('results.png')
```

# Урок 2

```python
from numpy.random import exponential
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axes = plt.subplots()
axes.plot(exponential(5, 20))
axes.set_title('График')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10,5))
for row, row_axes in enumerate(axes):
    for column, ax in enumerate(row_axes):
        ax.plot(exponential(column, 20))
        ax.set_title(f'Холост {row + 1, column + 1}')
fig.tight_layout()
fig.savefig('result2.png')
```
